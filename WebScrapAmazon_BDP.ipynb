```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from bs4 import BeautifulSoup
import requests
```

```python
def get_title(soup):
  try:
    titles = soup.find_all("span" , attrs = {"class":'a-size-medium a-color-base a-text-normal'})
    title = []
    for i in titles:
      title.append(i.text)
  except AttributeError:
    title = []
  return title


def get_price(soup):
  try:
    prices = soup.find_all("span" , attrs = {"class":'a-price-whole'})
    price = []
    for i in prices:
      price.append(i.text)
  except AttributeError:
    price = []
  return price

def get_rating(soup):
  try:
    ratings = soup.find_all("div" , attrs = {"class":'a-row a-size-small'})
    rating = []
    for i in ratings:
      rating.append(i.text)
    rc = [temp.split(" ")[0] for temp in rating]
  except AttributeError:
    rating = []
  return rc

def get_rating_count(soup):
  try:
    ratings = soup.find_all("div" , attrs = {"class":'a-row a-size-small'})
    rating = []
    for i in ratings:
      rating.append(i.text)
    rcc = [temp.split(" ")[5].replace(",","") for temp in rating]
  except AttributeError:
    rating = []
  return rcc

def get_mrp(soup):
  try:
    mrps = mrps = soup.find_all("span" , attrs = {"class":'a-price a-text-price'})
    mrp =[]
    for i in mrps:
      mrp.append(i.contents[0].text)
  except AttributeError:
    mrp = ""
  return mrp
```

```python
url = 'https://www.amazon.in/s?k=gaming&crid=3PPZ8DYG06OBC&sprefix=gami%2Caps%2C607&ref=nb_sb_noss_2'

webpage = requests.get(url)
webpage
```

```python
f = open("amazon.html", "wb")
f.write(webpage.content)
f.close()

f = open("amazon.txt", "wb")
f.write(webpage.content)
f.close()
```

```python
soup = BeautifulSoup(webpage.content, 'html.parser')
```

```python

title = get_title(soup)
price = get_price(soup)
mrp = get_mrp(soup)
ratings = get_rating(soup)
rc= get_rating_count(soup)

d = {'title': title , 'price': price , 'MRP': mrp, 'ratings':ratings , 'total_ratings': rc }
```

```python
df = pd.DataFrame(d)
df['title'].replace('', np.nan, inplace = True)
df = df.dropna(subset=['title'])
df.to_csv("amazon_data.csv", header=True, index= False)
```
